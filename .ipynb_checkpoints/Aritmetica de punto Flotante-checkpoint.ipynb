```python
from pylab import *
%matplotlib inline
```

```python
x = linspace(0,3*pi,100)
y = cos(x)*sin(2*x)
title("Gráfica de $\cos (x) \sin (x)$")
plot(x,y)
show()
```

```python
with xkcd():
    title("Gráfica de $\cos (x) \sin (x)$")
    plot(x,y)
    show()
```

```python
x = linspace(-4*pi,4*pi,100)
y = sin(x)/x
plot(x,y,label=r"$\frac{\sin(x)}{x}$")
title(u'Gráfica')
xlabel("$x$"),ylabel(r'$f(x)$')
legend()
show()
```

```python
figura = figure(figsize=(3,3),dpi=200)
x = linspace(-4*pi,4*pi,100)
y = sin(x)/x
plot(x,y,label=r"$\frac{\sin(x)}{x}$")
title(u'Gráfica')
xlabel("$x$"),ylabel(r'$f(x)$')
legend(),grid(True,linestyle=':')
show()
```

```python
fig, (ejes1,ejes2,ejes3) = subplots(3,1,sharex=True)

x1 = linspace(-3,3)
y1 = sin(x1)
x2 = linspace(-3,3)
y2 = cos(x2)
ejes1.plot(x1,y1,color='red')
ejes1.set_title("Graf 1")
ejes1.grid(True)
ejes2.plot(x2,y2)
show()
```

```python
a = range(1,7)
for ai in a:
    x = linspace(-2,10)
    y = (ai*x)**3+1./cosh(x)
    etiqueta = r"$\vec a_"+str(ai)+"$"
    plot(x,y,label=etiqueta)
legend()
show()
```

```python
theta = linspace(0,2*pi,400)
r = 2.*cos(6*theta)
r_circ = ones(len(theta))
with xkcd():
    ejespolares = subplot(111,projection='polar')
    ejespolares.plot(theta,r)
    ejespolares.plot(theta,r_circ)
    ejespolares.fill_between(theta,r,r_circ)
    show()
```

# Arimetica de punto flontes

## Fuentes de Error 

El error puede provenir de muchas fuentes al usar un método numérico:
  - Modelo / Error de datos
  - Error de truncamiento
  - Error de punto flotante
 
**Objetivo:** Clasificar y comprender cada tipo de error y explorar algunos enfoques simples para analizar el error.

## Modelos y error de datos

Errores en la formulación fundamental.
  - Lotka-Volterra - conejos fraccionarios, sin extinciones, etc.
  - Error de datos - Inexactitud en la medición o incertidumbres en los parámetros
 
Desafortunadamente, no podemos controlar el modelo y el error de datos directamente, pero podemos usar métodos que pueden ser más sólidos en la presencia de este tipo de errores.

## Error de truncamiento

Errores que surgen al aproximar una función con una función más simple, p. Ej. $ \sin(x) \approx x $ para $ | x | \approx 0 $. 

## Error de punto flotante

Errores que surgen al aproximar números reales con números de precisión finita y aritmética.
